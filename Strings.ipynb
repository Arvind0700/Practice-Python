```python
s= input('enter the string: ')
term=input('enter the term to be searched: ')
count=0
for i in s:
    if i==term:
        count+=1
print('The term',term,'appears',count,'times')        
```

```python
s= input('enter the string: ')
term=input('enter the term to be removed: ')

result=''
for i in s:
    if i!=term:
        result=result+i
print('String after removing',term, 'is:',result)
```

```python
s=input("Enter a string: ")
flag= True
for i in range(0, len(s)//2):
    if s[i]!=s[len(s)-1-i]:
        flag=False
        print("Not a palindrome")
        break
if flag:
    print("Palindrome")    
```

```python
s=input("Enter a string: ")
l=[]
temp=''
for i in s:
    if i!=' ':
        temp+=i
    else:
        l.append(temp)
        temp=''
l.append(temp)
print(l)        
```

```python
s=input("Enter a string: ")
l=[]
for i in s.split():
    l.append(i[0].upper()+i[1:].lower())
print(l)    
```

```python
number= int(input("Enter a number: "))

digits='0123456789'
result=''
while number!=0:
    result=digits[number%10]+result
    number=number//10
print("The number is:", result)
print(type(result))    
```
